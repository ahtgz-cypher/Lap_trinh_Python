```python
from math import *
```

```python
M = float(input("Enter M: "))
```

```python
denta_t = float(input("Enter denta t: "))
```

```python
C = 4.186
```

```python
J_to_kWh = 2.777e-7
```

```python
Q = M * C * denta_t
```

```python
cost_per_kWh = 8.9
```

```python
Q_kWh = Q * J_to_kWh
```

```python
cost = Q_kWh * cost_per_kWh
```

```python
print("Nang luong can thiet J: {:.2f}".format(Q))
```

```python
print("Nang luong can thiet kWh: {:.6f}".format(Q_kWh))
```

```python
print("Cost: {:.6f}".format(cost))
```
