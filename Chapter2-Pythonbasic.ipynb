```python
#Bài 31
def caesar_cipher(text, shift=3):
    result = ""
    for ch in text:
        if ch.isupper():
            result += chr((ord(ch) - ord('A') + shift) % 26 + ord('A'))
        elif ch.islower():
            result += chr((ord(ch) - ord('a') + shift) % 26 + ord('a'))
        else:
            result += ch
    return result

message = input("Nhập tin nhắn: ")
encrypted = caesar_cipher(message)
print("Tin nhắn đã được mã hóa:", encrypted)
```

```python
#Bài 32
def caesar_cipher(text, shift, mode="encrypt"):
    if mode == "decrypt":
        shift = -shift

    result = ""
    for ch in text:
        if ch.isupper():
            result += chr((ord(ch) - ord('A') + shift) % 26 + ord('A'))
        elif ch.islower():
            result += chr((ord(ch) - ord('a') + shift) % 26 + ord('a'))
        else:
            result += ch
    return result

message = input("Nhập tin nhắn: ")
shift = int(input("Nhập số ký tự dịch chuyển: "))
mode = input("Chọn chế độ (encrypt/decrypt): ").strip().lower()

if mode not in ["encrypt", "decrypt"]:
    print("Chế độ không hợp lệ!")
else:
    result = caesar_cipher(message, shift, mode)
    print("Kết quả:", result)
```

```python
#Bài 42
def factorial(n):
    if n < 0:
        return None
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

num = int(input("Nhập một số nguyên không âm: "))
fact = factorial(num)

if fact is not None:
    print(f"Giai thừa của {num} là: {fact}")
else:
    print("Số không hợp lệ. Vui lòng nhập số nguyên không âm.")
```

```python
#Bài 49
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Nhập một số nguyên: "))

if is_prime(num):
    print(f"{num} là số nguyên tố.")
else:
    print(f"{num} không phải là số nguyên tố.")
```

```python
#Bài 50
import random
def generate_password():
    length = random.randint(7, 10)
    password = ""
    for _ in range(length):
        char = chr(random.randint(33, 126))
        password += char
    return password

pwd = generate_password()
print("Mật khẩu ngẫu nhiên được tạo:", pwd)
```

```python
#Bài 51
def is_strong_password(password):
    if len(password) < 8:
        return False

    has_upper = False
    has_lower = False
    has_digit = False

    for ch in password:
        if ch.isupper():
            has_upper = True
        elif ch.islower():
            has_lower = True
        elif ch.isdigit():
            has_digit = True

    return has_upper and has_lower and has_digit

pwd = input("Nhập mật khẩu: ")

if is_strong_password(pwd):
    print("Mật khẩu của bạn là mật khẩu tốt.")
else:
    print("Mật khẩu của bạn không đủ mạnh.")
```

```python
#Bài 52
def is_perfect(n):
    if n <= 0:
        return False
    divisors_sum = 0
    for i in range(1, n):
        if n % i == 0:
            divisors_sum += i
    return divisors_sum == n

print("Các số hoàn hảo từ 1 đến 10000 là:")

for num in range(1, 10001):
    if is_perfect(num):
        print(num)

```

```python
#Bài 53
def all_sublists(lst):
    sublists = []
    n = len(lst)
    for i in range(n):
        for j in range(i, n):
            sublists.append(lst[i:j+1])
    sublists.append([])
    return sublists

my_list = [1, 2, 3]
result = all_sublists(my_list)
print("Các danh sách con của", my_list, "là:")
for sub in result:
    print(sub)
```

```python
#Bài 54
def format_words(words):
    if not words:
        return ""
    elif len(words) == 1:
        return words[0]
    else:
        return ", ".join(words[:-1]) + " and " + words[-1]

n = int(input("Nhập số lượng từ: "))
words = []

for i in range(n):
    word = input(f"Nhập từ thứ {i+1}: ").strip()
    words.append(word)

formatted = format_words(words)
print("Danh sách đã định dạng:")
print(formatted)
```

```python
#Bài 55
import zipfile
import os

def compress_files(zip_name, files):
    with zipfile.ZipFile(zip_name, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for file in files:
            if os.path.exists(file):
                zipf.write(file)
                print(f"Đã thêm {file} vào {zip_name}")
            else:
                print(f"{file} không tồn tại!")

def extract_files(zip_name, extract_dir='.'):
    with zipfile.ZipFile(zip_name, 'r') as zipf:
        zipf.extractall(extract_dir)
        print(f"Đã giải nén {zip_name} vào thư mục {extract_dir}")

print("1. Nén tệp")
print("2. Giải nén tệp")
choice = input("Chọn chức năng (1 hoặc 2): ")

if choice == '1':
    zip_name = input("Tên file zip muốn tạo (vd: archive.zip): ")
    files = input("Nhập tên các tệp muốn nén, cách nhau bởi dấu phẩy: ").split(',')
    files = [f.strip() for f in files]
    compress_files(zip_name, files)
elif choice == '2':
    zip_name = input("Nhập tên file zip cần giải nén: ")
    extract_dir = input("Nhập thư mục giải nén (mặc định .): ").strip() or '.'
    extract_files(zip_name, extract_dir)
else:
    print("Lựa chọn không hợp lệ.")
```
