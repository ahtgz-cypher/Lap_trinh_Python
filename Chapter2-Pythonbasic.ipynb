```python
# Bai 6
s = input("Enter the string: ")
s_upper = 0
s_lower = 0
for c in s:
    if c.isupper():
        s_upper+=1
    elif c.islower():
        s_lower+=1
    else:
        pass
print("Sum char upper case: ", s_upper)
print("Sum char lower case: ", s_lower)
```

```python
# Bai 7
num = int(input("Enter a number: "))
level = int(input("Enter level of num: "))
sum = 0
temp = num
while temp > 0:
    digit = temp % 10
    sum = sum + digit ** level
    temp//=10
if num == sum:
    print("This is a numer AmStrong")
else:
    print("This is not a number AmStrong")
    
```

```python
# Bai 10
total = 0
count = 0
while True:
    n = input("Enter a number: ")
    if n == "done":
        break
    value = float(n)
    total = total + value
    count = count + 1
average = total / count
print("Average: ", average)

```

```python
# Bai 11
numlist = list()
while True:
    inp = input("Enter a number: ")
    if inp == "done":
        break
    value = float(inp)
    numlist.append(value)

average = sum(numlist) / len(numlist)
print("Average: ", average)
print(numlist)
```

```python
# Bai 17
s = input("Enter string: ")
d = {"alphabet":0, "digit":0}
for ch in s:
    if ch.isalpha() == True:
        d["alphabet"] += 1
    elif ch.isdigit() == True:
        d["digit"] += 1
    else:
        pass
print("Number of alphabet: ",d["alphabet"])
print("Number of digit: ", d["digit"])
```

```python
# Bai 18
n = input("Enter the string: ")
d = {"in_hoa":0, "chu_thuong":0}
for i in n:
    if i.isupper():
        d["in_hoa"] += 1
    elif i.lower():
        d["chu_thuong"] += 1
    else:
        pass
print("Number of in hoa: ", d["in_hoa"])
print("Number of chu thuong: ", d["chu_thuong"])
```

```python
# Bai 19
ftext = open("romeo.txt")
tu_dien_cac_tu = {}
for dong in ftext:
    danh_sach_tu = dong.split()
    for tu in danh_sach_tu:
    tu_dien_cac_tu[tu] = tu_dien_cac_tu.get(tu, 0) + 1

danh_sach = []
for key, val in tu_dien_cac_tu.items():
    newtup = (val, key)
    danh_sach.append(newtup)

danh_sach = sorted(danh_sach, reverse=True)

for val, key in danh_sach[:10]:
    print(key, val)
```
